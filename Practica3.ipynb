```python
from gapminder import gapminder
import seaborn.objects as so

datosAmerica= gapminder[gapminder["continent"]=="Americas"]
(
    so.Plot( data= datosAmerica, x="year", y="lifeExp")
    .add(so.Lines( color="#bfca"), group="country").add(so.Line(), so.PolyFit(1))
)
```

```python
from gapminder import gapminder
import seaborn.objects as so


datosAsia= gapminder[gapminder["continent"]=="Asia"]
datosAmerica= gapminder[gapminder["continent"]=="Americas"]
datosEurope= gapminder[gapminder["continent"]=="Europe"]
datosAfrica= gapminder[gapminder["continent"]=="Africa"]
datosOceania= gapminder[gapminder["continent"]=="Oceania"]

data=[datosAsia, datosEurope, datosAfrica, datosAmerica, datosOceania]

for elem in data:
    display(
        so.Plot( data= elem, x="year", y="lifeExp")
        .add(so.Lines( color="#b0ca"), group="country").add(so.Line(), so.PolyFit(1))
    )
```
